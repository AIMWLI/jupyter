# 行内与独行
- 行内公式：将公式插入到本行内，符号：$公式内容$，如：$xyz$
- 独行公式：将公式插入到新的一行内，并且居中，符号：$$公式内容$$，如：$$xyz$$
# 上标、下标与组合
- 上标符号，符号：^，如：$x^4$
- 下标符号，符号：_，如：$x_1$
- 组合符号，符号：{}，如：${16}_{8}O{2+}_{2}$
